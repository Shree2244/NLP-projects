```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
import bs4 as bs
import urllib.request
import re
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
scraped_data = urllib.request.urlopen('https://en.wikipedia.org/wiki/Sports_betting')
article = scraped_data.read()
print("Raw data scraped from the article: ",scraped_data)
```

```python
parsed_article = bs.BeautifulSoup(article,'lxml')
print("Parsed article is: ",parsed_article)
```

```python
paragraphs = parsed_article.find_all('p')
print("Paragraph is: ",paragraphs)
```

```python
article_text = ""
for p in paragraphs:
    article_text += p.text
    print(p)

print("Article text: ",article_text)
```

```python
# Removing Square Brackets and Extra Spaces
article_text = re.sub(r'\[[0-9]*\]', ' ', article_text)
article_text = re.sub(r'\s+', ' ', article_text)
print("Removing square brackets and extra spaces:- \nThe text after using regex is: ",article_text)
# Removing special characters and digits
formatted_article_text = re.sub('[^a-zA-Z]', ' ', article_text )
formatted_article_text = re.sub(r'\s+', ' ', formatted_article_text)
print("Removing special characters and digits:- \nThe text after using regex is: ",formatted_article_text)
```

```python
sentence_list = nltk.sent_tokenize(article_text)
print("Text after sentence tokenize is ",article_text)
```

```python
stopwords = nltk.corpus.stopwords.words('english')
print("Set of stopword: ", stopwords)
```

```python
word_frequencies = {}
for word in word_tokenize(formatted_article_text):
    if word not in stopwords:
        if word not in word_frequencies.keys():
            word_frequencies[word] = 1
        else:
            word_frequencies[word] += 1
maximum_frequency = max(word_frequencies.values())
print("Word frequencies of the words: ",word_frequencies.items)
print("The maximum number of times any word has occured is: ",maximum_frequency)
```

```python
for word in word_frequencies.keys():
    word_frequencies[word] = (word_frequencies[word]/maximum_frequency)
print("Weighted freqeuncy of each word is: ", word_frequencies.items)
```

```python
sentence_scores = {}
for sent in sentence_list:
    for word in word_tokenize(sent.lower()):
        if word in word_frequencies.keys():
            if len(sent.split(' ')) < 50:
                if sent not in sentence_scores.keys():
                    sentence_scores[sent] = word_frequencies[word]
                else:
                    sentence_scores[sent] += word_frequencies[word]
print("Sentence scores: ", sentence_scores.items)
```

```python
import heapq
summary_sentences = heapq.nlargest(7, sentence_scores, key=sentence_scores.get)
("Summarised sentences before joining: ",summary_sentences)
```

```python
summary = ' '.join(summary_sentences)
print("Final summary is: ",summary)
```
